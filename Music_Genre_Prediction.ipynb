```python
#test and train model

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df=pd.read_csv("music.csv")
x=df.drop(columns=['genre'])
y=df['genre']

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(x_train, y_train)

prediction=model.predict(x_test)
score= accuracy_score(y_test, prediction)
score
```

```python
#generate descision tree

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df=pd.read_csv("music.csv")
x=df.drop(columns=["genre"])
y=df['genre']

model = DecisionTreeClassifier()
model.fit(x, y)

tree.export_graphviz(model, out_file='music-rec.dot',
                     feature_names=['age','gender'],
                     class_names=sorted(y.unique()),#displaying the class in each nodes
                     label='all', #every node has lables we can read
                     rounded=True, #rounded corners
                     filled=True) #descision tree each block filled with colour

with open("music-rec.dot", "r") as f:
    content = f.read()
print(content)

```
